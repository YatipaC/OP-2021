### Question 01:
(1 point) Prove that if functions $f(.), g(.)$ are convex, then $h(x)=\max \{f(x), g(x)\}$ is convex as well.

##### Sotution: 


The convex function are:
- Definition (Convex Function): A function $f: \mathbb{R}^{n} \rightarrow \mathbb{R}$ is convex, if for every $x$, $y \in \mathbb{R}^{n}$ and $0 \leq \lambda \leq 1$ the inequality
$$
f(\lambda x+(1-\lambda) y) \leq \lambda f(x)+(1-\lambda) f(y)
$$
holds.
The function $f$ is strictly convex, if for every $x \neq y \in \mathbb{R}^{n}$ and $0<\lambda<1$ the inequality
$$
f(\lambda x+(1-\lambda) y)<\lambda f(x)+(1-\lambda) f(y)
$$
holds. That is, the inequality defining the convexity of a function is strict whenever possible.
- A function f is convex, if its Hessian is everywhere positive semi-definite. This allows us to test whether a given function is convex.

So that
\begin{align*} 
h(x) &= \max \{f(x), g(x)\} \\
h(\lambda x+(1-\lambda) y) &= \max \{f(\lambda x+(1-\lambda) y), g(\lambda x+(1-\lambda) y)\} \\
h(\lambda x+(1-\lambda) y) &\leq \max \{ \lambda f(x)+(1-\lambda) f(y),  \lambda g(x)+(1-\lambda) g(y)\} \\
h(\lambda x+(1-\lambda) y) &\leq \lambda h(x)+(1-\lambda) h(y)   \\
\end{align*} 
As a result, h(x) is convex function

### Question 05:
(4 points) Solve the problem
$$
\min 2 x_{1}^{4}+3 x_{2}^{4}+2 x_{1}^{2}+4 x_{2}^{2}+x_{1} x_{2}-3 x_{1}-2 x_{2}, \quad \mathbf{x} \in \mathbb{R}^{2}
$$
by means of the gradient method with the Armijo inexact line search setting $\alpha=0.1, \gamma=0.9, \bar{t}=1$, and starting from the point $(0,0)$. \{Use $\|\nabla f(\mathbf{x})\|<10^{-3}$ as stopping criterion $\}$

Gradient: 
\begin{align*} 
8 x_{1}^{3}+12 x_{2}^{3}+4 x_{1}+8 x_{2}+x_{1} +x_{2}-5
\end{align*} 


```python
import numpy as np
import math 
import matplotlib.pyplot as plt
from numpy import linalg as LA
import warnings
warnings.filterwarnings("ignore")
```

```python
def obj_fn(x): 
    fn = 2*x[0]**4 + 3*x[1]**4 + 2*x[0]**2 + 4*x[1]**2 + x[0]*x[1] - 3*x[0] - 2*x[1]
    return fn

def grad_obj_fn(x): 
    fn = 8*x[0]**3 + 12*x[1]**3 + 4*x[0] + 8*x[1] + x[0] + x[1] - 3 - 2  
    return fn


## ver 1
def armijio(x):
#     alpha = 0.1 # if we implement algorithm from another formula
    c1 = np.arange(0.01,0.3,0.001)  
    beta = np.arange(0.01,0.8,0.002)  
    max_iter = 10000 
    gamma = 0.9
    i = 0 
    for i in range(1,max_iter): 
        alpha = gamma*beta[i]        
        if obj_fn(x) - c1[i]*alpha*LA.norm(grad_obj_fn(x))**2 <= obj_fn(x - alpha * grad_obj_fn(x)):
            break
    return alpha
    
```

```python

def gradient_descent(x0,tolerance):  
    x_k = x0 
    num_steps = 0
    x_l = []
    error = 1e-3
    while error > tolerance:
        gamma_i = armijio(x_k) 
        x_prev = x_k
        x_k = x_k - gamma_i*grad_obj_fn(x_k)
        gfk = grad_obj_fn(x_k) 
        x_l.append((x_k))
        num_steps += 1
        error = np.linalg.norm(x_k - x_prev)
    return num_steps, obj_fn(x_k), x_k,x_l,gamma_i    

 
def plot_feasible():
    X = np.arange(-1, 1, 0.1)
    Y = np.arange(-1, 1, 0.1)
    X, Y = np.meshgrid(X, Y)
    fv = np.vectorize(obj_fn, signature='(2)->()') # apply x and y coordinate into Mishra function
    Z = fv(np.stack((X, Y), axis=-1))
    fig, ax = plt.subplots(figsize= (10,10))  
    plt.contourf(X, Y, Z)  
    plt.grid(True) 
    plt.title("Trail of gradient descent method")  
    plt.xlabel('x')
    plt.ylabel('y')
   

x0 = np.array([0,0])
tolerance = 1e-4
plot_feasible()
num_steps_res,f_res, x_k_res,x_l_res,gamma_i_res = gradient_descent(x0,tolerance)
print(f"Final value: {x_k_res}\nGamma: {gamma_i_res}\nStep number: {num_steps_res}\n")
plt.plot(np.array(x_l_res)[:,0], np.array(x_l_res)[:,0], 'o-', color = 'w', label = "initial point=(0.05, 0.05, 0.05, 0.05, 0.05)")
plt.legend()
plt.plot()
```

```python
import numpy as np
import math 
import matplotlib.pyplot as plt
from numpy import linalg as LA
import warnings
warnings.filterwarnings("ignore")
```

```python
def hilmat(n, m=0): 
    if n == 1 and (m == 0 or m == 1):
        return np.array([[1]])
    elif m == 0:
        m = n
    v = np.arange(1, n + 1) + np.arange(0, m)[:, np.newaxis]
    return 1. / v


def obj_fn6(x):
    A = hilmat(len(x))
    fn = np.matmul(np.matmul(x.T, A),x) #x.T * A * x
    return fn

def grad_obj_fn6(x): 
    A = hilmat(len(x))
    fn = 2*np.matmul(A,x) #2 * A * x
    return fn

def plot_feasible6():
    X = np.arange(-2, 2, 0.1)
    Y = np.arange(-2, 2, 0.1) 
    X, Y = np.meshgrid(X, Y)
    fv = np.vectorize(obj_fn6, signature='(2)->()') # apply x and y coordinate into Mishra function
    Z = fv(np.stack((X, Y), axis=-1))
    fig, ax = plt.subplots(figsize= (10,10))  
    plt.contourf(X, Y, Z)
    plt.grid(True) 
    plt.title("Trail of gradient descent method")  
    plt.xlabel('x')
    plt.ylabel('y')
    
# Define x0
x0 = np.array([0.8, 0.8])
tolerance = 1e-4
```

### gradient_descent_const 

```python
 
def gradient_descent_const(x0,tolerance):
    x_k = x0
    i = 0 
    f_l =[]
    num_steps = 0
    x_l = []
    L = 100
    alpha = 1/L 
    error = 1e-3
    while error > tolerance:
        x_k = x_k - alpha*grad_obj_fn6(x_k)
        gfk = grad_obj_fn6(x_k) 
        x_l.append((x_k))
        num_steps += 1
        error = np.linalg.norm(gfk, ord=2) 
        
    return num_steps, obj_fn6(x_k), x_k,x_l 

num_steps_res,f_res, x_k_res,x_l_res = gradient_descent_const(x0,tolerance)
print(f"optimal x: {x_k_res}\nMinimal result: {f_res}\nStep number: {num_steps_res}\n")

  
plot_feasible6() 
plt.plot(np.array(x_l_res)[:,0], np.array(x_l_res)[:,0], 'o-', color = 'w', label = f"initial point=({x0})")
plt.legend()
plt.plot()
```

### Steepest descent


According to $$\alpha^{k}=\operatorname{argmin} f\left(x^{k}-\alpha \nabla f\left(x^{k}\right)\right)=\phi(\alpha)$$
So That
\begin{align*}
g^k &= 2Ax\\
\phi'(\alpha) = \nabla(x^k - \alpha g^k)(-g^k) &= 0\\
2A(x^k - \alpha g^k)^T(-g^k) &= 0\\
x^{kT} g^k -  \alpha g^{kT} g^k &= 0\\ 
\alpha &= \frac{x^{kT} g^k}{g^{kT} g^k}\\
\end{align*}

```python
 
def steepest_descent(x0,tolerance): 
    error = 1e-3
    A = hilmat(len(x0))  
    x_k = x0 
    num_steps = 0
    x_l = []
    while error > tolerance:
        num_steps += 1
        grad_terms = grad_obj_fn6(x_k) 
        alpha = (x_k.T@grad_terms)/(grad_obj_fn6(x_k.T) @ grad_terms)  
        x_k = x_k - alpha*grad_terms
        x_l.append(x_k)  
        error = np.linalg.norm(grad_terms, ord=2) 
        
    return num_steps, obj_fn6(x_k), x_k, x_l   

 
num_steps_res,f_res, x_k_res, x_l_res = steepest_descent(x0,tolerance)
print(f"optimal x: {x_k_res}\nMinimal result: {f_res}\nStep number: {num_steps_res}\n")

  
plot_feasible6() 
plt.plot(np.array(x_l_res)[:,0], np.array(x_l_res)[:,0], 'o-', color = 'w', label = f"initial point=({x0})")
plt.legend()
plt.plot()
```

### Heavy ball method (constant parameters)

$$
x^{k+1}=x^{k}-\alpha \nabla f\left(x^{k}\right)+\beta\left(x^{k}-x^{k-1}\right)
$$
- $0 \leq \beta<1, \quad 0<\alpha<\frac{2(1+\beta)}{L}$
- $\alpha=\frac{4}{(\sqrt{L}+\sqrt{m})^{2}}, \beta=\left(\frac{\sqrt{L}-\sqrt{m}}{\sqrt{L}+\sqrt{m}}\right)^{2}$ when $\mu$ is known

- Convergence rate (near minimum):
$$
\left\|x^{k+1}-x^{*}\right\| \approx c\left(\frac{\sqrt{\varkappa}-1}{\sqrt{\varkappa}+1}\right)^{k}\left\|x^{0}-x^{*}\right\|
$$

```python
def heavy_ball(xk, tolerance, lipschitz, bk, ak):
    x = [xk, xk]
    xk_l = []
    k = 1
    error = 1e3
    errors = []
    while error > tolerance:
        #print(error)
        xk = x[-1]
        _xk = x[-2]

        gfk = grad_obj_fn6(xk)
        alpha = ak * 2 * (1 + bk) / lipschitz
        beta = bk

        xk = xk - alpha * gfk + beta * (xk - _xk)
        xk_l.append(xk)
        x.append(xk)

        error = np.linalg.norm(gfk, ord=2)
        errors.append(error)

        k += 1

    return xk, k - 1, errors,xk_l
search = {} 
grad = grad_obj_fn6(x0)
lipschitz = np.linalg.norm(grad, ord=2)
for ak in np.linspace(0.4, 0.6, 10):
    for bk in np.linspace(0., 0.99, 115):
        xk, k, errors_hb,xk_l_res = heavy_ball(x0, tolerance, lipschitz, bk, ak)
        search.update({k: [xk, bk, ak, errors_hb]})

min_k = min(list(search.keys()))
xk = search[min_k][0]
bk = search[min_k][1]
ak = search[min_k][2]
errors_hb = search[min_k][3]
k_hb = min_k


print(f"x optimzal: { np.round(xk, 4)}\nMinimal result: {obj_fn6(xk)}\nStep number: {min_k}\n")
print("\nOptimal beta =", np.round(bk, 4))
print("Optimal ak=", np.round(ak, 4))
print("Optimal alpha =", np.round(ak * 2 * (1 + bk) / lipschitz, 4))
print("error = %.4f" % errors_hb[-1])
_, _, _,xk_l_res = heavy_ball(x0, tolerance, lipschitz, np.round(bk, 4), np.round(ak, 4))
plot_feasible6() 
plt.plot(np.array(xk_l_res)[:,0], np.array(xk_l_res)[:,0], 'o-', color = 'w', label = f"initial point=({x0})")
plt.legend()
plt.plot()
```

```python
def conjucate_gradient(x0,tolerance): # grad_obj_fn6     np.linalg.norm(gfk, ord=2)
    xk = x0
    A = hilmat(len(xk))
    gfk = np.dot(A, xk)  #grad_obj_fn6(xk) # g^{k}
    pk = -gfk ## define p^{k=0}
    err = np.linalg.norm(gfk, ord=2) 
    x_l = []
    num_steps = 0
    while err > tolerance:
        apk = np.dot(A, pk) #  A p^{k}
        gfk2 = np.dot(gfk, gfk)
        
        alpha = gfk2 / np.dot(apk, pk) # \alpha_{k} =||g^{k}||^{2}/(A p^{k}, p^{k})
        xk = xk + alpha * pk # x^{k+1} = x^{k}+\alpha_{k} p^{k}
        gfk = gfk + alpha * apk # g^{k+1} = g^{k}+\alpha_{k} A p^{k}
        beta = np.dot(gfk, apk) / np.dot(apk, pk) # \frac{\left\|g^{k+1}\right\|^{2}}{\left\|g^{k}\right\|^{2}} 
        pk = -gfk + beta * pk
        
        num_steps += 1
        err = np.linalg.norm(gfk) 
        x_l.append(xk)
    return num_steps, obj_fn6(x_l[-1]), x_l[-1], x_l   


num_steps_res,f_res, x_k_res, x_l_res = conjucate_gradient(x0,tolerance)
print(f"optimal x: {x_k_res}\nMinimal result: {f_res}\nStep number: {num_steps_res}\n")

  
plot_feasible6() 
plt.plot(np.array(x_l_res)[:,0], np.array(x_l_res)[:,0], 'o-', color = 'w', label = f"initial point=({x0})")
plt.legend()
plt.plot()
```

### Fast gradient method
##### Nesterovs fast gradient method
- $\left\{\begin{array}{l}\text { If } k \geq 1: \text { choose } \theta_{k} \in(0,1) \text { so that } \frac{\left(1-\theta_{k}\right) t_{k}}{\theta_{k}^{2}} \leq \frac{t_{k-1}}{\theta_{k-1}^{2}} \\ y=\left(1-\theta_{k}\right) x_{k}+\theta_{k} v_{k} \\ x_{k+1}=y-t_{k} \nabla f(y) \\ v_{k+1}=x_{k}+\frac{1}{\theta_{k}}\left(x_{k+1}-x_{k}\right)\end{array}\right.$
- constant step size $t_{k}=t \in(0,1 / L]$

```python
def fast_grad(x0,tolerance):
    x_k = x0
    theta_k = 1
    v_k = x_k
    t_k = 1/lipschitz
    err = 1e-3 
    num_step = 0
    x_l = []
    while err > tolerance:
        theta_prev = theta_k
        for theta_k in np.linspace(0.01, 0.99, num=100):
            if (1 - theta_k) / theta_k ** 2 <= 1 / theta_prev ** 2:
                break
        y = (1 - theta_k) * x_k + theta_k * v_k 
        xk_prev = x_k
        x_k = y - t_k * grad_obj_fn6(y)
        x_l.append(x_k)
        v_k = xk_prev + 1 / theta_k * (x_k - xk_prev)
        err = np.linalg.norm(grad_obj_fn6(y), ord=2)  
        num_step += 1
    return num_step,obj_fn(x_k), x_k, x_l

num_steps_res,f_res, x_k_res, x_l_res = fast_grad(x0,tolerance)
print(f"optimal x: {x_k_res}\nMinimal result: {f_res}\nStep number: {num_steps_res}\n")

  
plot_feasible6() 
plt.plot(np.array(x_l_res)[:,0], np.array(x_l_res)[:,0], 'o-', color = 'w', label = f"initial point=({x0})")
plt.legend()
plt.plot()
```
